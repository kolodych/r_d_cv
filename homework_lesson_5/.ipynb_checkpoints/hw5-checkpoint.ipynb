```python
import cv2
import math
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]

# Load image
img = cv2.imread('farby1.jpg')
# Convert it to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.subplot(221), plt.imshow(img), plt.title('Original')

# Black, dark gray, light gray, white
colors_BGW = np.array([[0, 0, 0],
                   [64, 64, 64],
                   [192, 192, 192],
                   [255, 255, 255]])




def quant(image, colors):
    # Apply quantization
    # Prepare for quantization
    # Cast the image to float
    img_tmp = np.copy(image)
    img_tmp = img_tmp.astype(np.float32)
    rows, cols, chnl = img_tmp.shape
    quantized = np.zeros_like(img_tmp)

    for r in range(rows):
        for c in range(cols):
            # Extract the original pixel value
            pixel = img[r, c]
            # Find the closest colour from the pallette (using absolute value/Euclidean distance)
            # Note: You may need more than one line of code here
            euclide = np.sqrt(np.sum((pixel-colors)**2,axis=1))
        
            new_pixel = colors[np.argmin(euclide)]
            # Apply quantization
            quantized[r, c, :] = new_pixel
   

    quantized = quantized.astype(np.uint8)
    return quantized



             
    
# Compute PSNR
def psnr(origin, compressed):
    mse = np.mean((origin - compressed) ** 2) 
    print("MSE", mse)
    psnr = 10 * math.log10(255.0**2 / mse)  
    return psnr 

def fsd(image, colors):
    # Make a temporal copy of the original image, we will need it for error diffusion
    img_tmp = np.copy(image)
    img_tmp = img_tmp.astype(np.float32)
    dithering = np.zeros_like(img_tmp)
    rows, cols, chnl = img_tmp.shape
    for r in range(1, rows-1):
        for c in range(1, cols-1):
            # Extract the original pixel value
            pixel = img_tmp[r, c]
            # Find the closest colour from the pallette (using absolute value/Euclidean distance)
            # Note: You may need more than one line of code here
            euclide = np.sqrt(np.sum((colors-pixel)**2,axis=1))
        
            new_pixel = colors[np.argmin(euclide)]      
        
            # Compute quantization error
            quant_error = pixel-new_pixel
        
            # Diffuse the quantization error accroding to the FS diffusion matrix
            # Note: You may need more than one line of code here
            img_tmp[r][c] = new_pixel
            img_tmp[r + 1] [c    ] = img_tmp [r + 1] [c    ] + quant_error * 7 / 16
            img_tmp[r - 1] [c + 1] = img_tmp [r - 1] [c + 1] + quant_error * 3 / 16
            img_tmp[r    ] [c + 1] = img_tmp [r    ] [c + 1] + quant_error * 5 / 16
            img_tmp[r + 1] [c + 1] = img_tmp [r + 1] [c + 1] + quant_error * 1 / 16
            
            # Apply dithering
            dithering[r, c, :] = new_pixel

    dithering = dithering.astype(np.uint8) 
    return dithering


optimal = quant(img, colors_BGW)
dithering = fsd(img, colors_BGW)
print("Optimal PSNR", psnr(img, optimal))
print("FSD PSNR", psnr(img, dithering))
plt.subplot(222), plt.imshow(optimal), plt.title('Quantized')
plt.subplot(223), plt.imshow(dithering), plt.title('Dithering')
```

```python
colors = np.array([[0, 0, 0],
                   [255, 255, 255]])

optimal = quant(img, colors)
dithering = fsd(img, colors)

print("Optimal PSNR", psnr(img, optimal))
print("FSD PSNR", psnr(img, dithering))

plt.subplot(221), plt.imshow(img), plt.title('Original')
plt.subplot(222), plt.imshow(optimal), plt.title('Quantized B/W')
plt.subplot(223), plt.imshow(dithering), plt.title('Dithering B/W')
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=16).fit(np.reshape(img, (-1, 3)))
colors = kmeans.cluster_centers_

dithering = fsd(img, colors)

plt.subplot(221), plt.imshow(dithering), plt.title('Dithering 16 colors')

kmeans = KMeans(n_clusters=32).fit(np.reshape(img, (-1, 3)))
colors = kmeans.cluster_centers_
dithering = fsd(img, colors)

plt.subplot(222), plt.imshow(dithering), plt.title('Dithering 32 colors')

kmeans = KMeans(n_clusters=256).fit(np.reshape(img, (-1, 3)))
colors = kmeans.cluster_centers_

dithering = fsd(img, colors)
plt.subplot(223), plt.imshow(dithering), plt.title('Dithering 256 colors')



```
